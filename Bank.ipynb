### Which customer profiles (job, education, etc.) have higher subscription rates?
### Use value_counts(normalize=True) within each group to get the proportion of “yes”
### and “no” responses.

```python
r=df.groupby('job')['y'].value_counts(normalize=True).unstack()
r
```

```python
t=df.groupby('education')['y'].value_counts(normalize=True).unstack()
t
```

```python
r.plot(kind='bar',stacked=True,figsize=[10,5],color=['skyblue','burlywood'],edgecolor='black')
plt.legend(title='subscription',bbox_to_anchor=(1.0,1.0))
plt.title('Subscription Rates by Job')
plt.ylabel('Percentage')
plt.show()
      
```

# Subscription Rates by Job
### Students have the highest subscription rate.
### Retired customers follow closely behind.
### Other job groups (e.g., blue-collar, services, management, entrepreneur, housemaid, technician)
### generally show low subscription rates, with most not subscribing.


```python
t.plot(kind='bar',stacked=True,figsize=[10,5],color=['skyblue','burlywood'],edgecolor='black')
plt.legend(title='subscription',bbox_to_anchor=(1.0,1.0))
plt.title('Subscription Rates by Education')
plt.ylabel('Percentage')
plt.show()
```

# Subscription Rates by Education:
### Education level is a strong predictor of subscription likelihood.

# Previous Campaign Impact:
### Does having prior contact with a client increase the chance of success?

```python
dd=df.groupby('poutcome')['y'].value_counts(normalize=True).unstack()*100
dd
```

```python
dd.plot(kind='bar',figsize=[10,5],color=['skyblue','burlywood'],edgecolor='black')
plt.legend(title='subcription',bbox_to_anchor=(1.0,1.0))
plt.grid(axis='y', linestyle='--')
plt.title('Success Rate in Previous Campaign ')
plt.ylabel('Percentage')
plt.show()
```

# Comparison of Success Rates (by Subscription Status):
### Failure: About 85% did not subscribe, and only 15% subscribed, showing low success.
### Nonexistent: Nearly 90% did not subscribe, making this the least responsive group.
### Success: Around 65% subscribed, indicating a high chance of repeat success.

### Subscription rates increase from 2 to 6 contacts, with the highest success around 5 contacts.

# Temporal Analysis:
### Which months have higher subscription rates?

```python
s=df.groupby('month')['y'].value_counts(normalize=True).unstack()*100
month_order = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
s=s.reindex(month_order)
s
```

```python
plt.figure(figsize=(12,6))
plt.plot(s.index, s[1], marker='o', color='burlywood', linewidth=2)
plt.title('Trend of Subscription Rate Over Months')
plt.xlabel('Month')
plt.ylabel('Subscription Rate(%)')
plt.grid(linestyle='--')
plt.xticks(rotation=45)
plt.show()
```

# Trend of Subscription Rate Over Months:
### March, September, October, and December have significantly higher subscription rates.
### Other months show much lower subscription rates.

### Is there any effect of the day of the week on subscription success?

```python
day=df.groupby('day_of_week')['y'].value_counts(normalize=True).unstack()*100
weekday_order = ['mon','tue','wed','thu','fri']
day = day.reindex(weekday_order)
```

```python
day.plot(kind='bar',figsize=[10,5],color=['skyblue','burlywood'],edgecolor='black')
plt.title('Subscription Rate by Day of the Week')
plt.xlabel('Day of Week')
plt.ylabel('Percentage')
plt.legend(title='subcription',bbox_to_anchor=(1.0,1.0))
plt.grid(axis='y', linestyle='--')
plt.show()
```

# Subscription Rate by Day of the Week:
### Subscription success shows minimal variation by day of the week.
### Rates remain in a narrow range of about 10–12% across all weekdays,
### with Tuesday to Thursday performing slightly better than Monday and Friday.
